```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
from scipy.stats import pearsonr
import squarify
```

#### Lendo os Dados 

```python
df = pd.read_csv('excel_ma.csv', sep = ';', decimal = ',', encoding = 'latin-1')
df = df.iloc[:,3:]
```

```python
#eixo X: variáveis de performance
#eixo Y: variáveis de morfologia

vars_morf = ['Comprimento Médio da Fibra  (mm)', 'Coarseness (mg/100m)','Nº de Fibras  (g x 106)', 
             'Finos (%)', 'Vasos   (Nº/grama)','Área Média do Vaso (mm²)', 'Curl  (%)', 'Kink (mm)',
             'Diâmetro da Fibra  (mm)', 'Espessura da Parede (µm)','Diâmetro do Lúmen (mm)', 'Fração Parede (%)']

vars_perform = ['Umidade', 'Gramatura','Viscosidade', 'pH da Polpa para Torres 1 e 2', 'PRESSÃO DO SECADOR',
                'VELOCIDADE TELA MV QCS', 'CONSUMO ESPECÍFICO DE VAPOR','Abertura da Válvula de Vapor p/ Secador', 
                'TEMP. 2 DO SECADOR','Pressao VBP para MC2', 'PRODUÇÃO BRANQUEAMENTO']
```

### Gráfico X vs. Y e valor de correlação de Pearson

```python
def grafico_e_corr(x,y):

    fig, ax1 = plt.subplots()

    x = x
    y = y

    print('Coeficiente de Correlação de Pearson:',round(pearsonr(x,y)[0],3))

    ax1.plot(x, y, 'bo')

    plt.show()
```

```python
for i in vars_perform:
    for j in vars_morf:
        print('X:',i, 'Y:',j)
        grafico_e_corr(df[i].values, df[j].values)
```

### Mapa de Calor: correlação entre todas as variáveis

```python
#mapa de calor

k = 23
corrmat = df.corr()
cols = corrmat.nlargest(k, 'CONSUMO ESPECÍFICO DE VAPOR')['CONSUMO ESPECÍFICO DE VAPOR'].index
f, ax = plt.subplots(figsize=(15, 12))
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=0.75)
s = sns.heatmap(cm,  linewidths=0.18,cbar=True, annot=True, square = True, fmt='.2f', 
                annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
ax.set_ylim(0,23)
s.set_title("Mapa de calor de correlação entre as variáveis", size=12)
plt.show()

plt.savefig('heatmap.png')
```

### Tree Map para cada variável de performance vs. variáveis de morfologia

```python
def corr(x,y):

    x = x
    y = y
    
    resp = pearsonr(x,y)[0]
    
    return round(resp,3)
```

```python
all_corr = []

for i in vars_perform:
    list_corr = []
    
    for j in vars_morf:
        correlation = corr(df[i].values,df[j].values)
        list_corr.append(correlation)
        
    all_corr.append(list_corr)
```

```python
for index in range(len(vars_perform)):
    
    print(vars_perform[index],'\nCorrelações de Pearson com parâmetros de performance:\n')
    print(list(zip(vars_morf,all_corr[index])))
    
    labels = [vars_morf[i] + ':\n' + str(all_corr[index][i]) for i in range(len(vars_morf))]

    color_list = ['#0f7216', '#b2790c', '#ffe9a3','#f9d4d4', '#d35158', '#ea3033']

    plt.figure(figsize = (15,15))

    plt.rc('font', size=8)

    correlation = [i if i > 0 else i*(-1.0) for i in all_corr[index]]

    squarify.plot(sizes=correlation, label=labels,color=color_list, alpha=0.7)

    plt.axis('off')

    plt.show()
```
